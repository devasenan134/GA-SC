```python
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
ga_path = 'pickles/ga/'
kbga_path = 'pickles/kbga/'
```

```python
ga_files = ['n_run_az_100fc.pkl', 'n_run_imdb_100fc.pkl', 'n_run_yelp_100fc.pkl']
kbga_files = ['n_run_az_kbga_nokbps.pkl', 'n_run_az_kbga_es50.pkl', 'n_run_az_kbga_cmp.pkl', 'n_run_az_kbga_vp.pkl']
```

## GA Comparison

```python
with open(ga_path+ga_files[0], 'rb') as gf:
    data = pickle.load(gf)
    x = range(1, 101)
    for n in data:
        y = n[1]    
        plt.plot(x, y)
    plt.title("GAb - Amazon | 100Feature count")
    plt.xlabel('No. of Generations')
    plt.ylabel('Accuracy')
    plt.show()
```
