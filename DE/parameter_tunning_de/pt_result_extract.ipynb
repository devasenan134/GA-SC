```python
def plot_graph(avg_scores, pt=2):
    plt.figure(figsize=(15, 9))
    x = range(1, 101)
    legends = []
    for i in range(len(avg_scores)):
        y = avg_scores[i][0]   
        plt.plot(x, y)
        legends.append(avg_scores[i][pt])
    plt.title(f"DE - Amazon | F Score - {avg_scores[0][1]}| Crossover probability - PT")
    plt.xlabel('Generations')
    plt.ylabel('Accuarcy')
    plt.legend(legends)
    plt.show()  
```
