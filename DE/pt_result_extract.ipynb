```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle as pk
```

```python
file = open('n_run_az_de_100fc_pt_avg.pkl', 'rb')
cp = pk.load(file)
```

```python
cp[0]
```

```python
def plot_graph(avg_scores):
    plt.figure(figsize=(15, 9))
    x = range(1, 101)
    legends = []
    for i in range(len(avg_scores)):
        y = avg_scores[i][0]   
        plt.plot(x, y)
        legends.append(avg_scores[i][1])
    plt.title(f"DE - Amazon | Crossover probability - PT")
    plt.xlabel('Generations')
    plt.ylabel('Accuarcy')
    plt.legend(legends)
    plt.show()  
```

```python
plot_graph(cp[0])
```
