```python
def plot_graph(avg_scores):
    x = range(1, 101)
    mt_pb = []
    for i in range(len(avg_scores)):
        y = avg_scores[i][0]   
        mt_pb.append(avg_scores[i][2])
        plt.plot(x, y)
    plt.title(f"GA - Amazon | Crossover probability - {avg_scores[i][1]}")
    plt.xlabel('Generations')
    plt.ylabel('Accuarcy')
    plt.legend(mt_pb)
    plt.show()  
```
